```python
from django.http import JsonResponse
from mysql.connector  import Error, connect
import pandas
```

```python
import os
os.environ.setdefault("DJANGO_SETTINGS_MODULE", "ai4dlab.settings")

import django
django.setup()

from django.core.management import call_command
```

```python
columns =['id','patient_id','age','country','district','region','main_complaint','history_of_present_illness',
        'review_of_other_systems','past_medical_history','gynaecological_history','family_social_history','dietary_history',
        'general_examination','local_examination','systemic_examination','provisional_diagnosis','differential_diagnosis',
        'final_diagnosis','radiological','laboratory','doctors_remarks','medicines','treatment_regime','recommendation',
        'created_at','updated_at','facility','user_id']
```

```python
def db_connect(host, port, database,user,password):
    try:
        connection = connect(host=host, port=port, database=database, user=user, password=password)
        if connection.is_connected():
            cursor = connection.cursor()
            cursor.execute("SELECT * FROM dportal_patientdata;")
            records = cursor.fetchall()
            if len(records) != 0:
                df = pandas.DataFrame(records, columns=columns)
                return df
    except Error as e:
        print("Error: ", e)
        return None
```
